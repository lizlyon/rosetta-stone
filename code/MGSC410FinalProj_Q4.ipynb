```python
def load_rosetta_stone_data():
    """
    This function loads the rosetta stone data
    """

    import pandas as pd
    rosetta_stone_data = pd.read_csv(
        "https://github.com/connoralydon/rosetta_stone_final/raw/main/data/rosetta_stone_clean.csv")

    rosetta_stone_data["language"] = rosetta_stone_data["language"].astype(
        "category")
    rosetta_stone_data["subscription_event_type"] = rosetta_stone_data["subscription_event_type"].astype(
        "category")
    rosetta_stone_data["purchase_store"] = rosetta_stone_data["purchase_store"].astype(
        "category")
    rosetta_stone_data["country"] = rosetta_stone_data["country"].astype(
        "category")
    rosetta_stone_data["user_type"] = rosetta_stone_data["user_type"].astype(
        "category")

    rosetta_stone_data["subscription_start_date"] = pd.to_datetime(
        rosetta_stone_data["subscription_start_date"])

    rosetta_stone_data["demo_user"] = rosetta_stone_data["demo_user"].astype(
        "bool")
    rosetta_stone_data["free_trial_user"] = rosetta_stone_data["free_trial_user"].astype(
        "bool")
    rosetta_stone_data["auto_renew"] = rosetta_stone_data["auto_renew"].astype(
        "bool")
    rosetta_stone_data["email_subscriber"] = rosetta_stone_data["email_subscriber"].astype(
        "bool")
    rosetta_stone_data["push_notifications"] = rosetta_stone_data["push_notifications"].astype(
        "bool")
    rosetta_stone_data["subscription_type_limited"] = rosetta_stone_data["subscription_type_limited"].astype(
        "uint8")
    rosetta_stone_data["subscription_type_lifetime"] = rosetta_stone_data["subscription_type_lifetime"].astype(
        "uint8")

    # rosetta_stone_data["send_count"] = rosetta_stone_data["send_count"].astype("int64")
    # rosetta_stone_data["unique_open_count"] = rosetta_stone_data["unique_open_count"].astype("int64")
    # rosetta_stone_data["unique_click_count"] = rosetta_stone_data["unique_click_count"].astype("int64")
    # rosetta_stone_data["purchase_amount_usd"] = rosetta_stone_data["purchase_amount_usd"].astype("int64")
    # rosetta_stone_data["subscription_length_days"] = rosetta_stone_data["subscription_length_days"].astype("int64")
    # rosetta_stone_data["total_app_interactions"] = rosetta_stone_data["total_app_interactions"].astype("int64")
    # rosetta_stone_data["launch_app_interactions"] = rosetta_stone_data["launch_app_interactions"].astype("int64")
    # rosetta_stone_data["purchase_amount_usd_imputed"] = rosetta_stone_data["purchase_amount_usd_imputed"].astype("int64")
    return rosetta_stone_data


rosetta_stone_data = load_rosetta_stone_data()
```

```python
rosetta_stone_data.to_csv("rosetta_stone_data.csv")
```

```python
features = ['send_count', 'unique_open_count', 'unique_click_count',
            'launch_app_interactions', 'purchase_amount_usd_imputed',
            'language_ALL', 'language_ARA', 'language_CHI', 'language_DAR',
            'language_DEU', 'language_EBR', 'language_ENG', 'language_ESC',
            'language_ESP', 'language_FAR', 'language_FRA', 'language_GLE',
            'language_GRK', 'language_HEB', 'language_HIN', 'language_IND',
            'language_ITA', 'language_JPN', 'language_KIS', 'language_KOR',
            'language_LAT', 'language_NED', 'language_PAS', 'language_POL',
            'language_POR', 'language_RUS', 'language_SVE', 'language_TGL',
            'language_TUR', 'language_URD', 'language_VIE',
            'subscription_event_type_RENEWAL', 'purchase_store_Web',
            'demo_user_True', 'free_trial_user_True', 'country_Other',
            'country_US/Canada', 'user_type_Other', 'lead_platform_Unknown',
            'lead_platform_Web', 'email_subscriber_True', 'push_notifications_True', 'subscription_length_days']
```

```python
import os
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'

score_val = []
for k in range(2, 10):
    init_model = KMeans(n_clusters = k)
    init_model.fit(rs_data[features])
    predictions = init_model.predict(rs_data[features])
    score_val.append(metrics.silhouette_score(rs_data[features], predictions))

K = np.argmax(score_val) + 2
print(K, score_val[np.argmax(score_val)])

model = KMeans(n_clusters = 2)
model.fit(rs_data[features])
rs_data['KM'] = model.predict(rs_data[features])
model_score = metrics.silhouette_score(rs_data[features], rs_data['KM'])
print(model_score)
```
